```python
x=np.linspace(0,2*pi,500)
```

```python
y=np.sin(x)
z=np.cos(x)
```

```python
y
```

```python
x
```

```python
plt.xlabel("Angles")
plt.ylabel("Sin(x)")
plt.title("Sin Graph")
plt.plot(x,y,color="violet",label="Sin(x)")
plt.plot(x,z,'g',label="Cos(x)")
plt.legend()
```

```python
plt.plot(x,y,'r-.',x,z,'purple')
```
